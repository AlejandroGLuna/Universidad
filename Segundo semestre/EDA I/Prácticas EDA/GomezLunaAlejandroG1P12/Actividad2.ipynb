```python
def funcion1(L):
    if L==[]:
        return 1
    elif len(L)==1:
        return L[0]
    else:
        return L[0]*funcion1(L[1:])

lista1=[1,2,3,4,5,6,7,8]
lista2=funcion1(lista1)
print(lista2)
```

```python
def funcion3(L,n):
    if n==1:
        return L[0]
    else:
        return L[0]+funcion3(L[1:],n-1)
    
lista1=[1,2,3,4,5,6,7,8]
lista2=funcion3(lista1,8)
print(lista2)
```

```python
def move(n,x,y,z):
    if n==1:
        print('move',x,'to',y)
    else:
        move(n-1,x,z,y)
        print('move',x,'to',y)
        move(n-1,z,y,x)
        
move(10,"A","B","C")
```

```python
def mystery(S):
    N=S.split()
    print(N)
    N=''.join(N)
    print(N)
    if len(N)==1 or len(N)==0:
        return True
    else:
        if N[0]==N[-1] and mystery(N[1:-1]):
            return True
        else:
            return False

test=input("Ingrese una cadena \n")
print(mystery(test))
```
